```python
from datetime import timedelta, date

date_strings = []

def daterange(start_date, end_date):
    for n in range(int ((end_date - start_date).days)):
        yield start_date + timedelta(n)

start_date = date(2017, 4, 1)
end_date = date(2017, 4, 10)
for single_date in daterange(start_date, end_date):
    date_strings.append(single_date.strftime("%Y-%m-%d"))
```

```python
links = []
for date_str in date_strings:
    page_url = date_url.format(date_str, 1, 1)
    print(date_str)
    with urllib.request.urlopen(page_url) as fp:
        soup = bs(fp.read(), 'lxml')
    links += [host+link['href'] for link in soup.select('.listRight li h2 a')]
    print('done!')
```
